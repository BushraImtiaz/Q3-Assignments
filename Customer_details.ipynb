```
class Employee:
    working_hours = 12
    def __init__(self, ID ,name ,salary ,depart, Whour):
        self.aID = ID
        self.aname = name
        self.asalary = salary
        self.adepart = depart
        self.awhour = Whour
    def printdetails(self):
        return f"ID : {self.aID}\nName : {self.aname}\nsalary : {self.asalary}\nDepartment : {self.adepart}\nWorking Hours : {self.awhour}\n"
    # Using Decorator changing class's variable value
    @classmethod
    def changeWorkingHour(cls, newHours):
        cls.NewWorkingHour = newHours
class Programmer(Employee):
    def __init__(self, ID, name, salary, depart, Whour, languages):
        self.aID = ID
        self.aname = name
        self.asalary = salary
        self.adepart = depart
        self.awhour = Whour
        self.languages = languages
    def printProgDetails(self):
        return f"ID : {self.aID}\nName : {self.aname}\nsalary : {self.asalary}\nDepartment : {self.adepart}\n" \
               f"Working Hours : {self.awhour}\nLanguages : {self.languages}\n"
def writeDetailintoFile(string):
    f = open("EmployeeData.txt", "a")  # "r+" is used to read and write both function
    f.write(string)  # f.write() function return number of character used in used
    f.write("----------------------------------------------\n")
    f.close()
emp1 = Programmer("AD00001", "Bushra Imtiaz", 35500, "AI Developer", 10, ["Python","C#","Asp.Net","MVC.Net"])
emp2 = Programmer("AD00002", "Babloo Imtiaz", 55500, "Website designer", 9, ["Python","C#","Asp.Net","MVC.Net"])
emp3 = Employee("AD00003", "Ayehsa", 34500, "Data Operator",10)
emp4 = Employee("AD00004", "Muhammad Arham", 54500, "Database Administrator",10)
print("-------------- Print Details using Inheritance-----------------")
print(emp1.printProgDetails())
print(emp2.printProgDetails())
print("-------------- Print Details Super Class-----------------")
print(emp3.printdetails())
print(emp4.printdetails())
writeDetailintoFile(emp1.printProgDetails())
writeDetailintoFile(emp2.printProgDetails())
writeDetailintoFile(emp3.printdetails())
writeDetailintoFile(emp4.printdetails())
```
